```python
def one_for_all(df):
    
    cols = df.columns
    new_cols = []
    rows = df.shape[0]
    
    for col in cols:
        new_col = ""
        for letter in col:
            if letter == " ":
                letter = "_"
                new_col += letter
            else:
                new_col += letter
        new_cols.append(new_col.lower())
    df.columns = new_cols
    
    for col in new_cols:
        na = df[col].isna().sum()
        if na > (25*rows/100):
            df = df.drop([col], axis=1)
        else: df = df[df[col].isna()==False]
    
    num = pd.DataFrame(df.select_dtypes(np.number))
    cat = pd.DataFrame(df.select_dtypes("object"))
    
    return df, num, cat

df1, num, cat = one_for_all(df1)
```

```python
# Stars: w/f, sm, ir
# Number: hits
# Inches: height
# Pounds: weight
# Money: value, wage, release_clause
# Time: joined
```

```python
def starless(df):
    
    values = []
    
    for value in df:
        value = value.replace(' ★', "")
        value = value.replace('★', "")
        value = int(value)
        values.append(value)
    
    df = values

    return df

df1["w/f"] = starless(df1["w/f"])
df1["sm"] = starless(df1["sm"])
df1["ir"] = starless(df1["ir"])
```

```python
def show_me_the_money(df):
    
    values = []
    
    for value in df:
        value = value.replace('€', "")
        
        if "." in value:
            split = value.split(".")
            value = split[0] + split[1]
            value = value.replace("M", "00000")
            value = value.replace("K", "00")

        else:       
            value = value.replace("M", "000000")
            value = value.replace("K", "000")
            
        value = int(value)
        values.append(value)
    
    df = values

    return df

df1["value"] = show_me_the_money(df1["value"])
df1["wage"] = show_me_the_money(df1["wage"])
df1["release_clause"] = show_me_the_money(df1["release_clause"])
df1["hits"] = show_me_the_money(df1["hits"])
```

```python
def inches_to_cm(x):
    
    values = []
    
    for value in x:
        value = value.replace('"', "")
        split = value.split("'")
        split[0] = float(split[0])
        split[1] = float(split[1])
        split[0] = split[0]*30.48
        split[1] = split[1]*2.54
        value = split[0]+split[1]
        values.append(value)
    
    return values

df1["height"] = inches_to_cm(df1["height"])
df1["height"]
```

```python
df1["joined"] = pd.to_datetime(df1["joined"], errors="coerce")
df1["joined"]
```

```python
df1["weight"] = df1["weight"].map(lambda x: x.rstrip('lbs'))
df1["weight"] = pd.to_numeric(df1["weight"], errors='coerce')
```
